```python
import pandas as pd
import plotly.express as px
```

```python
pokemon = pd.read_csv('Pokemon Data - National Pokedex.csv')
pokemon
```

```python
pokemon_tidy = pokemon.drop(columns = ['Type II',\
                                       'Ability II',\
                                       'Hidden Ability',\
                                       'EV Worth',\
                                       'Gender',\
                                       'Egg Group I',\
                                       'Egg Group II',\
                                       'Evolve']).dropna()
pokemon_tidy
```

```python
pokemon_sorted = pokemon_tidy.sort_values(by = 'Catch', ascending = True)
pokemon_sorted
pokemon_sorted.head(15)
```

```python
fig = px.scatter(pokemon_sorted, x = 'Catch', y = 'Total',
                 color = 'Type I')
fig.show()
```

```python
fig = px.scatter(pokemon_sorted, x = 'Catch', y = 'Total',
                 color = 'Type I',
                 title = 'Pokemon Stats Organized by Type',
                size = 'HP')
fig.show()
```
